```
from khaiii import KhaiiiApi

khaiii = KhaiiiApi()
for word in khaiii.analyze('어릴때보고 지금다시봐도 재밌어요ㅋㅋ'):
    print('\n{}'.format(word))
    for w in word.morphs:
        print('- {} / {}'.format(w.lex, w.tag))
```
